```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

from ortools.sat.python import cp_model
import math
```

```python
#model creation
model = cp_model.CpModel()

#max no of table possible
num=7

#varible 

Jaime = model.NewIntVar(1, num, 'Jaime')
Beth = model.NewIntVar(1, num, 'Beth')
Jon = model.NewIntVar(1, num, 'Jon')
Daenerys=model.NewIntVar(1, num, 'Daenerys')
Cersei=model.NewIntVar(1, num, 'Cersei')
Sansa=model.NewIntVar(1, num, 'Sansa')
Tyrion=model.NewIntVar(1, num, 'Tyrion')

# Creates the constraints.

model.Add(Jaime==Cersei) #friends
model.Add(Beth==Tyrion) #friends
model.Add(Jon==Daenerys) #friends
model.Add(Jon!=Jaime) #Enemies
model.Add(Daenerys!=Cersei) #Enemies
model.Add(Jon!=Tyrion) #Enemies
model.Add(Sansa!=Cersei) #Enemies

 # Creates a solver and solves the model.
solver = cp_model.CpSolver()
status = solver.Solve(model)

print(status)
```

```python
table=[]

table.append(solver.Value(Jaime))
table.append(solver.Value(Beth))
table.append(solver.Value(Jon))
table.append(solver.Value(Cersei))
table.append(solver.Value(Daenerys))
table.append(solver.Value(Sansa))
table.append(solver.Value(Tyrion))
```

```python
table=list(dict.fromkeys(table))
```

```python
if status == cp_model.FEASIBLE:
    print('Jaime table no.= %i' % solver.Value(Jaime))
    print('Beth table no.= %i' % solver.Value(Beth))
    print('Jon table no.= %i' % solver.Value(Jon))
    print('Cersei table no.= %i' % solver.Value(Cersei))
    print('Daenerys table no. = %i' % solver.Value(Daenerys))
    print('Sansa table no.= %i' % solver.Value(Sansa))
    print('Tyrion table no.= %i' % solver.Value(Tyrion))
```

```python
print("total number of table required:", len(table))
```

# Printing all the possible solution

```python
class VarArraySolutionPrinter(cp_model.CpSolverSolutionCallback):
    """Print intermediate solutions."""

    def __init__(self, variables):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.__variables = variables
        self.__solution_count = 0

    def on_solution_callback(self):
        self.__solution_count += 1
        for v in self.__variables:
            print('%s=%i' % (v, self.Value(v)), end=' ')
        print()

    def solution_count(self):
        return self.__solution_count
```

```python
solver = cp_model.CpSolver()
solution_printer = VarArraySolutionPrinter([Jaime, Beth, Jon,Cersei,Daenerys,Sansa,Tyrion])
status = solver.SearchForAllSolutions(model, solution_printer)
```
