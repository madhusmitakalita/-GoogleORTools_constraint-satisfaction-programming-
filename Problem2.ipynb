```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
from ortools.sat.python import cp_model

import pandas as pd
import numpy as np


```

```python
#changing column names while importing data
df = pd.read_csv('Course_Student_Association_Matrix.csv',
                 names=name1)
```

```python
#remove the two last column as it stores NaN value
df = df.iloc[:, :-1]
df = df.iloc[:, :-1]
```

```python
df
```

```python
time_Slot=4
num_students=500
num_subject=20
max_num_days=60
rooms=6

room_Capacity=50

all_num_students = range(num_students)
all_time_Slot = range(time_Slot)
all_rooms= range(rooms)
all_subject = range(num_subject)
all_days=range(max_num_days)
```

```python
#creating the model & variable creation
model= cp_model.CpModel()

S={}

for i in range(num_subject):
    for j in range(max_num_days):
             for k in range(time_Slot):
                for l in range(rooms):
                    S[(i,j,k,l)] = model.NewBoolVar("x(%d,%d,%d,%d)" % (i,j,k,l))
    
```

```python
#Exam conflict: 2 exams that share students must not occur in the same period. (hard constraint)
for j in range(max_num_days):
        for k in range(time_Slot):
              for s in range (num_students):
                        a=[]
                        for c in range(num_subject):
                            if (df.iloc[(c,s)]==1):
                                a.append(c)
                        model.Add(sum(S[(i,j,k,l)] for i in a for l in range(rooms))<= 1)
```

```python
# single course should be schedule only once
for i in range(num_subject):
    ex=sum(S[(i,j,k,l)] for j in range(max_num_days) for k in range(time_Slot) for l in range(rooms))
    model.Add(ex==1)
```

```python
student_per_course=np.count_nonzero(df, axis=1)

#Room capacity: A room's seating capacity must suffice at all times.

for j in range(max_num_days):
        for k in range(time_Slot):
            for l in range(rooms):
                for i in range (num_subject):
                    if (student_per_course[i]>50):#check the room capacity
                        model.Add(S[(i,j,k,l)]==0)
                        
                
            
```

```python
#Shared conflict: Exam should not have to share its room with any other exam
for l in range(rooms):
    for j in range(max_num_days):
             for k in range(time_Slot):
                model.Add(sum(S[(i,j,k,l)] for i in range (num_subject))<= 1)
            
  
```

```python
#Exam number limit constraint: The same student should not have more than 2 exams per day. (hard constraint)
for s in range (num_students):
    a=[]
    for c in range(num_subject):
        if (df.iloc[(c,s)]==1):
            a.append(c)
    for j in range(max_num_days):
                model.Add(sum(S[(i,j,k,l)] for i in a for l in range(rooms) for k in range(time_Slot))<= 2)
   
```

```python
#No consecutive exam constraint: The same student should not have 2 exams in a row
for s in range (num_students):
    a=[]
    for c in range(num_subject):
        if (df.iloc[(c,s)]==1):
            a.append(c)
    for j in range(max_num_days):
        sum2=0
        for p in range(time_Slot):
            for k in range(p,p+1):# p & its next consecutive slot p+1
                for i in a:
                    b=sum(S[(i,j,k,l)] for l in range(rooms))
                    sum2=b+sum2
        model.Add(sum2<2)
        
             
         
            
```

```python
solver = cp_model.CpSolver()
status = solver.Solve(model)
print(status)
```

```python
day1=[]
t=["10:00 AM - 12:00PM","1:00 PM - 3:00PM","3:00 PM - 5:00PM","5:00 PM - 7:00PM"]
for i in range(num_subject):
    for j in range(max_num_days):
             for k in range(time_Slot):
                for l in range(rooms):
                        if(solver.Value(S[(i,j,k,l)])==1):
                            day1.append(j)
                            print ("course:",i+1,"day",j+1,"time slot",t[(k)],'in room',l+1)
                        
```

```python
days1= list(dict.fromkeys(day1))# remove the duplicate
print("total number of days to schedule without soft constraint:",len(days1))
```

```python

#soft constrain

#The number of days required for scheduling all the exams should be as minimum as possible(soft constraint)

days=[]
for i in range(num_subject):
    for j in range(max_num_days):
             for k in range(time_Slot):
                for l in range(rooms):
                    if S[(i,j,k,l)]==1:#if its schedule
                        days.append(j)#append the day
days= list(dict.fromkeys(days))# remove the duplicate
model.Minimize(len(days))#minimize the count of days
```

```python
solver = cp_model.CpSolver()
status = solver.Solve(model)
print(status)
```

```python
results=[]
days2=[]
for i in range(num_subject):
    for j in range(max_num_days):
             for k in range(time_Slot):
                for l in range(rooms):
                        if(solver.Value(S[(i,j,k,l)])==1):
                            results.append([i+1,j+1,t[(k)],l+1])
                            days2.append(j)
                            print ("course:",i+1,"day",j+1,"time slot",t[(k)],'in room',l+1)
```

```python

#importing the results to CSV
import csv

Results1=pd.DataFrame(results)
Results1.columns = ["Course", "Day", "Time_slot", "Room_no"]
Results1.to_csv('Results.csv',index=False)
```
