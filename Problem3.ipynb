```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
from ortools.sat.python import cp_model

import pandas as pd
import random
import numpy as np
from random import randint

import math
```

```python
# data provided
Total_PPE=700
total_hospital=5
no_of_Patients=[30,20,35,34,15]
no_of_Staff=[50,55,65,72,45]

n_patients=sum(no_of_Patients)
n_staff=sum(no_of_Staff)

patients_severity={}
patients_severity_hospital={}


# Initialize severity level for each patients randomly in the range of 1-5
for i in range(total_hospital):
    e=0
    for j in range(no_of_Patients[i]):
        patients_severity[i,j] = randint(1, 5)
        f=patients_severity[i,j]
        e+=f
    # getting the severity density of the hospital
    patients_severity_hospital[i]=e/no_of_Patients[i]
        
```

```python
model = cp_model.CpModel()
x={}

# to declare maxium number of PPE a staff can own 

max=math.ceil(Total_PPE/n_staff)

#Creating the variable
for i in range(total_hospital):
    for j in range(no_of_Staff[i]):
        x[(i, j)] = model.NewIntVar(0, max, "%d,%d" % (i,j))
```

```python
x
```

```python
#All staff should be given at least one PPE
for i in range(total_hospital):
    for j in range(no_of_Staff[i]):
        model.Add( x[(i, j)] >= 1)
        
#No two staff can share a single PPE is already satisfied as the value is integer to be owned by staff
```

```python
#No PPE should remain un-allocated
s=0
for i in range(total_hospital):
    for j in range(no_of_Staff[i]):
        a=x[(i, j)]
        s+=a
model.Add(s==Total_PPE)
```

```python
#soft constraint

#taking the % of the severity of the hospital as gain factor
gain_severity_patients = 100/5
#taking the % of the staff count as gain factor
gain_severity_staff=100/n_staff


# Maximization objective
soft_csts=[]

for i in range(total_hospital):
    factor=gain_severity_staff*no_of_Staff[i]\
    +gain_severity_patients*patients_severity_hospital[i]
    factor=round(factor)
    print(factor)
    for j in range(no_of_Staff[i]):
        soft_csts.append(factor*x[(i,j)])
        
model.Maximize(sum(soft_csts))
```

```python
solver = cp_model.CpSolver()
solver.parameters.max_time_in_seconds = 60.0
status = solver.Solve(model)
```

```python
print(status)
```

```python
for i in range(total_hospital):
    print("hospital",i+1)
    print("total Staff",no_of_Staff[i])
    print("Severity density of hospital",patients_severity_hospital[i])
    a=0
    for j in range(no_of_Staff[i]):
        #print("\tStaff", solver.Value(x[(i,j)]))
        b=solver.Value(x[(i,j)])
        a+=b
    print("total PPE allocated",a)
    print("--------------")
```
